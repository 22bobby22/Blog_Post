# Which countries are more keen to recommend online courses

I will take the first 5 countries with more respondents in the survey, then see which countries have more percentage of respondents that recommend to take online courses.

```python
#import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#load the data
df = pd.read_csv('./survey_results_public.csv')
df.head()
```

```python
#get the first 5 countries with more respondents
grouped = df.groupby('Country').count().reset_index()
sortedvalues = grouped.sort_values('Respondent', ascending=False)
countries = sortedvalues['Country'].head()
```

```python
#get only the respondents with the countries selected
df2 = df.dropna(subset=['CousinEducation'])
df2 = df2[df2['Country'].isin(countries)]
df2
```

```python
#get the respondents that recommend online courses
df3 = df2[df2['CousinEducation'].str.contains('Take online courses')]
df3
```

```python
#count for each country how many respondents recommend online courses
onlinecount = df3.groupby('Country').size().reset_index(name='Count').sort_values('Count', ascending=False)
onlinecount
```

```python
#total of respondents for each country to calculate the percentages
totalcount = df2.groupby('Country').size().reset_index(name='Total').sort_values('Total', ascending=False)
totalcount
```

```python
#create a dataframe where we can see al the variables together
df4 = onlinecount
df4['Total'] = totalcount['Total']
df4['Percentage'] = df4['Count']/df4['Total']

#show the dataframe sorting by percentages
df4.sort_values('Percentage', ascending=False)
```

```python
#select the country and the percentages to create a plot
df5 = df4[['Country', 'Percentage']]
df5
```

```python
#create the plot
df5.plot(kind='bar', x='Country')
```

# Summary

Percentage of respondents that recommend online course:

- India: 65.7%
- United Kingdom: 65.1%

- Canada: 62.6%
